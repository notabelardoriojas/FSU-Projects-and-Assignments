```python
def a6():
    tVals = np.linspace(0,100,2001)
    a = 2.0
    b = 1.1
    xVals = np.array([])
    yVals = np.array([])
    for i in range(len(tVals)):
        t = tVals[i]
        xt = (a+b)*np.cos(t) - b * np.cos((((a/b)+1)*t))
        yt = (a+b)*np.sin(t) - b * np.sin((((a/b)+1)*t))
        xVals = np.append(xVals,xt)
        yVals = np.append(yVals,yt)
    plt.plot(xVals,yVals)
    plt.show()
        
a6()
```
